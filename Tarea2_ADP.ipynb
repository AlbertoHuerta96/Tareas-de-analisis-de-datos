```python
#Solucion a ejercicio 1

import numpy as np

vectores = [[0, 1, 0], [1, 1, 0], [1, 2, 3], [-1, 0, 1], [0, 0, 1], [-1, 1, 0], [2, 1, 1], [3, 0.5, -1]]
a = [1, 0, 0]

for v in vectores:
    producto_punto = v[0] * a[0] + v[1] * a[1] + v[2] * a[2]
    norma_v = np.sqrt(v[0]**2 + v[1]**2 + v[2]**2)
    norma_a = np.sqrt(a[0]**2 + a[1]**2 + a[2]**2)
    cos_theta = producto_punto / (norma_v * norma_a)
    if cos_theta > 1:
        cos_theta = 1
    if cos_theta < -1:
        cos_theta = -1
    angulo = np.arccos(cos_theta) * 180 / np.pi
    if angulo > 80:
        print("Vector:", v, "Ángulo:", angulo)
```

```python
#Solucion a ejercicio 2

import numpy as np
import matplotlib.pyplot as plt

n = 200
x = np.linspace(-6, 6, n)
y = np.linspace(-6, 6, n)
Z = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        x2_y2 = x[i]**2 + y[j]**2
        if x2_y2 <= 4:
            Z[j, i] = np.sin(x2_y2)
        elif x2_y2 < 12:
            Z[j, i] = np.cos(x[i] - y[j])
        else:
            Z[j, i] = -1

plt.imshow(Z, extent=[-6, 6, -6, 6], origin='lower', cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x, y)')
plt.show()
```

```python
#Solucion al ejercicio 4

import numpy as np

n = 400
datos = np.random.normal(10, 2, n)

print("Datos generados:")
for i in range(n):
    print(datos[i])

suma = 0
for x in datos:
    suma = suma + x
media = suma / n

suma_cuadrados = 0
for x in datos:
    suma_cuadrados = suma_cuadrados + (x - media)**2
desviacion = np.sqrt(suma_cuadrados / n)

contador = 0
for x in datos:
    if 8 <= x <= 12:
        contador = contador + 1

print("\nMedia de la muestra:", media)
print("Desviación estándar de la muestra:", desviacion)
print("Número de valores entre 8 y 12:", contador)
```

```python
#Solucion a ejercicio 5

import numpy as np

n = 200
categorias = ['A', 'B', 'C']
probabilidades = [0.5, 0.3, 0.2]
muestra = np.random.choice(categorias, n, p=probabilidades)

print("Datos generados:")
for i in range(n):
    print(muestra[i])

conteo_A = 0
conteo_B = 0
conteo_C = 0
for valor in muestra:
    if valor == 'A':
        conteo_A = conteo_A + 1
    elif valor == 'B':
        conteo_B = conteo_B + 1
    else:
        conteo_C = conteo_C + 1

print("\nNúmero de 'A':", conteo_A)
print("Número de 'B':", conteo_B)
print("Número de 'C':", conteo_C)
```

```python
#Solucion a ejercicio 6

import numpy as np
import matplotlib.pyplot as plt

n = 100
senal_base = np.full(n, 100)
ruido = np.random.normal(0, 5, n)
senal_ruidosa = senal_base + ruido

print("Señal ruidosa:")
for i in range(n):
    print(senal_ruidosa[i])

suma_cuadrados = 0
for i in range(n):
    suma_cuadrados = suma_cuadrados + (senal_ruidosa[i] - senal_base[i])**2
mse = suma_cuadrados / n

plt.plot(range(n), senal_ruidosa, label='Señal ruidosa')
plt.plot(range(n), senal_base, label='Señal original', linestyle='--')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('Señal con ruido gaussiano')
plt.legend()
plt.grid(True)
plt.show()

print("\nError medio cuadrático (MSE):", mse)
```
